bureau_balance Table

```python
b2= pd.read_csv("../data/raw/bureau_balance.csv").dropna()
b1= pd.read_csv("../data/raw/bureau.csv").dropna()
```

```python
b1.shape
```

```python
b2.shape
```

```python
overdue = ["1", "2", "3", "4", "5"]
b2["is_overdue"] = b2["STATUS"].isin(overdue).astype(int)
bb_agg = (
    b2
    .groupby("SK_ID_BUREAU")["is_overdue"]
    .sum()
    .reset_index()
    .rename(columns={"is_overdue": "bureau_overdue_months"})
)
```

```python
b = b1.merge(bb_agg, on="SK_ID_BUREAU", how="left")
b["bureau_overdue_months"] = b["bureau_overdue_months"].fillna(0)
```

```python
bureau_final = (
    b
    .groupby("SK_ID_CURR")["bureau_overdue_months"]
    .sum()
    .reset_index()
)
```

```python
ap = ap.merge(bureau_final, on="SK_ID_CURR", how="left")
ap["bureau_overdue_months"] = ap["bureau_overdue_months"].fillna(0)
```

Bureau Table

```python
b1= pd.read_csv("../data/raw/bureau.csv").dropna()
b_avg = (
    b1
    .groupby("SK_ID_CURR")["AMT_CREDIT_SUM"]
    .mean()
    .reset_index()
    .rename(columns={"AMT_CREDIT_SUM": "AVG_Loan_Amount"})
)
ap = ap.merge(b_avg, on="SK_ID_CURR", how="left")
```

```python
active_loans = (
    b1[b1["CREDIT_ACTIVE"] == "Active"]
    .groupby("SK_ID_CURR")
    .size()
    .reset_index(name="total_active_bureau_loans")
)
ap = ap.merge(active_loans, on="SK_ID_CURR", how="left")
```

```python
ap["AVG_Loan_Amount"] = ap["AVG_Loan_Amount"].fillna(0)
ap["total_active_bureau_loans"] = ap["total_active_bureau_loans"].fillna(0)
```

Installments payments Table

```python
inst = pd.read_csv("../data/raw/installments_payments.csv").dropna()
```

```python
inst["payment_delay"] = inst["DAYS_ENTRY_PAYMENT"] - inst["DAYS_INSTALMENT"]
inst["payment_rate"] = np.where(inst["AMT_INSTALMENT"] > 0,
                                    inst["AMT_PAYMENT"] / inst["AMT_INSTALMENT"],
                                    np.nan)
```

```python
payment_delay = inst.groupby("SK_ID_CURR")["payment_delay"].mean().reset_index()
payment_delay = payment_delay.rename(columns={"payment_delay": "inst_mean_payment_delay"})
ap = ap.merge(payment_delay, on="SK_ID_CURR", how="left")
ap["inst_mean_payment_delay"] = ap["inst_mean_payment_delay"].fillna(ap["inst_mean_payment_delay"].median())
```

```python
payment_rate= inst.groupby("SK_ID_CURR")["payment_rate"].mean().reset_index()
payment_rate = payment_rate.rename(columns={"payment_rate": "inst_mean_payment_rate"})
ap = ap.merge(payment_rate, on="SK_ID_CURR", how="left")
ap["inst_mean_payment_rate"] = ap["inst_mean_payment_rate"].fillna(ap["inst_mean_payment_rate"].median())
```

Previous Application Table

```python
prev = pd.read_csv("../data/raw/previous_application.csv")
```

```python
prev_avg_amt = (
    prev.groupby("SK_ID_CURR")["AMT_APPLICATION"]
    .mean()
    .reset_index()
    .rename(columns={"AMT_APPLICATION": "avg_prev_amt_requested"})
)
```

```python
prev["status"] = prev["NAME_CONTRACT_STATUS"].astype(str).str.lower()

prev["is_prev_approved"] = prev["status"].str.contains("approved", na=False).astype(int)

prev["is_prev_rejected"] = prev["status"].str.contains("refus|reject|canceled", na=False).astype(int)
```

```python
prev_counts = (
    prev.groupby("SK_ID_CURR")[["is_prev_approved", "is_prev_rejected"]]
    .sum()
    .reset_index()
    .rename(columns={
        "is_prev_approved": "prev_num_approved",
        "is_prev_rejected": "prev_num_rejected"
    })
)

prev_total = (
    prev.groupby("SK_ID_CURR")
    .size()
    .reset_index(name="prev_num_applications")
)

prev_feats = prev_avg_amt.merge(prev_counts, on="SK_ID_CURR", how="outer")
prev_feats = prev_feats.merge(prev_total, on="SK_ID_CURR", how="left")

prev_feats["avg_prev_amt_requested"] = prev_feats["avg_prev_amt_requested"].fillna(prev_feats["avg_prev_amt_requested"].median())
prev_feats["prev_num_approved"] = prev_feats["prev_num_approved"].fillna(0).astype(int)
prev_feats["prev_num_rejected"] = prev_feats["prev_num_rejected"].fillna(0).astype(int)
prev_feats["prev_num_applications"] = prev_feats["prev_num_applications"].fillna(0).astype(int)
```

```python
ap = ap.merge(prev_feats, on="SK_ID_CURR", how="left")
```

```python
ap["avg_prev_amt_requested"] = ap["avg_prev_amt_requested"].fillna(ap["avg_prev_amt_requested"].median())
ap["prev_num_approved"] = ap["prev_num_approved"].fillna(0).astype(int)
ap["prev_num_rejected"] = ap["prev_num_rejected"].fillna(0).astype(int)
ap["prev_num_applications"] = ap["prev_num_applications"].fillna(0).astype(int)
```

Credit Card Table

```python
cc = pd.read_csv("../data/raw/credit_card_balance.csv")
```

```python
cc_amt= (
    cc.groupby("SK_ID_CURR")["AMT_BALANCE"]
    .mean()
    .reset_index()
    .rename(columns={"AMT_BALANCE": "avg_cc_amt_balance"})
)
ap = ap.merge(cc_amt, on="SK_ID_CURR", how="left")
ap["avg_cc_amt_balance"] = ap["avg_cc_amt_balance"].fillna(ap["avg_cc_amt_balance"].median())
```

```python
cc_total= (
    cc.groupby("SK_ID_CURR")["AMT_PAYMENT_TOTAL_CURRENT"]
    .mean()
    .reset_index()
    .rename(columns={"AMT_PAYMENT_TOTAL_CURRENT": "avg_cc_amt_payment_total_current"})
)
ap = ap.merge(cc_total, on="SK_ID_CURR", how="left")
ap["avg_cc_amt_payment_total_current"] = ap["avg_cc_amt_payment_total_current"].fillna(ap["avg_cc_amt_payment_total_current"].median())
```

```python
cc_sk= (
    cc.groupby("SK_ID_CURR")["SK_DPD"]
    .max()
    .reset_index()
    .rename(columns={"SK_DPD": "max_cc_sk_dpd"})
)
ap = ap.merge(cc_sk, on="SK_ID_CURR", how="left")
ap["max_cc_sk_dpd"] = ap["max_cc_sk_dpd"].fillna(0)
```

```python
cc_sk= (
    cc.groupby("SK_ID_CURR")["SK_DPD_DEF"]
    .max()
    .reset_index()
    .rename(columns={"SK_DPD_DEF": "max_cc_sk_dpd_def"})
)
ap = ap.merge(cc_sk, on="SK_ID_CURR", how="left")
ap["max_cc_sk_dpd_def"] = ap["max_cc_sk_dpd_def"].fillna(0.0)
```

```python
cc_agg = (
        cc
        .groupby("SK_ID_CURR")
        .agg(
            avg_cc_max_limit_used=("AMT_CREDIT_LIMIT_ACTUAL", "max"),
            avg_cc_mean_drawings_atm_current=("AMT_DRAWINGS_ATM_CURRENT", "mean"),
        )
        .reset_index()
    )
ap = ap.merge(cc_agg, on="SK_ID_CURR", how="left")
ap["avg_cc_max_limit_used"] = ap["avg_cc_max_limit_used"].fillna
(ap["avg_cc_max_limit_used"].median())
ap["avg_cc_mean_drawings_atm_current"] = ap["avg_cc_mean_drawings_atm_current"].fillna
(ap["avg_cc_mean_drawings_atm_current"].median())
```
