Bureau Features

```python
b2= pd.read_csv("../data/raw/bureau_balance.csv")
b1= pd.read_csv("../data/raw/bureau.csv")
```

```python
b1.shape
```

```python
b2.shape
```

```python
overdue = ["1", "2", "3", "4", "5"]
b2["is_overdue"] = b2["STATUS"].isin(overdue).astype(int)
bb_agg = (
    b2
    .groupby("SK_ID_BUREAU")["is_overdue"]
    .sum()
    .reset_index()
    .rename(columns={"is_overdue": "bureau_overdue_months"})
)
```

```python
b1.head()
```

```python
b = b1.merge(bb_agg, on="SK_ID_BUREAU", how="left")
b["bureau_overdue_months"] = b["bureau_overdue_months"].fillna(0)
```

```python
bureau_final = (
    b
    .groupby("SK_ID_CURR")["bureau_overdue_months"]
    .sum()
    .reset_index()
)
```

```python
ap = ap.merge(bureau_final, on="SK_ID_CURR", how="left")
ap["bureau_overdue_months"] = ap["bureau_overdue_months"].fillna(0)
```
